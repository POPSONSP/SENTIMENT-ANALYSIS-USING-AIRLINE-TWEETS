### Text Data Preprocessing: text data is cleaned, preprocessed, and transformed into numerical features 

```python
# Define a function for processing text data
def process_text(text):
    # Convert the entire text to lowercase for consistency
    text = text.lower()

    # Remove numerical digits using a regular expression
    text = re.sub('[^A-Za-z]+', ' ', text)

    # Remove punctuation marks using string.punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenize the text into a list of words
    tokens = word_tokenize(text)

    # Remove common English stop words using NLTK's stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Perform stemming on each word using the Porter Stemmer
    porter = PorterStemmer()
    tokens = [porter.stem(word) for word in tokens]

    # Join the processed tokens back into a text string
    processed_text = ' '.join(tokens)

    # Return the cleaned and processed text
    return processed_text
```

```python
# Apply the text cleaning function to the 'text' column
data['cleaned_text'] = data['text'].apply(process_text)
```

### Lexicon-Based Sentiment Analysis using TextBlob: Polarity-based Classification
### Function to classify the sentiment of text into positive, negative, or neutral categories based on polarity scores calculated using TextBlob.


```python
# Define a function for sentiment analysis using TextBlob
def analyze_sentiment(text):
    # Create a TextBlob object for the input text
    analysis = TextBlob(text)
    
    # Get the sentiment polarity (-1 to 1)
    sentiment_polarity = analysis.sentiment.polarity
    
    # Classify the sentiment based on polarity
    if sentiment_polarity > 0:
        return "Positive"
    elif sentiment_polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Apply the sentiment analysis function to the 'text' column
data['sentiment'] = data['text'].apply(analyze_sentiment)
```

# Machine Learning Modelling of the sentiment outcome
### Objective: To utilize machine learning algorithms for modeling the sentiment outcomes related to airline experiences.
### Approach: Implement various machine learning algorithms to model and predict sentiment outcomes in the context of airline customer experiences.
### Methodology: Employ algorithms such as Random Forest, and Support Vector Machine to train models on labeled data for sentiment prediction.

### Data Partitioning: Function to split data into 80% training and 20% testing and encode labels

```python
# The feature variable is cleaned_text
# The label is the sentiment outcome
# Using a random_state = 0 to reproduce the same results later
def split_and_encode_data(text_data, labels):
    X_train, X_test, y_train, y_test = train_test_split(text_data, labels, test_size=0.2, random_state=0)
    encoder = preprocessing.LabelEncoder()
    train_y = encoder.fit_transform(y_train)
    test_y = encoder.fit_transform(y_test)
    return X_train, X_test, train_y, test_y
```

```python
# Split data and encode labels
X_train, X_test, y_train, y_test = split_and_encode_data(data['cleaned_text'], data['sentiment'])
```

# TF-IDF (Term Frequency-Inverse Document Frequency)
### Transform the raw text data into TF-IDF features. 
### TF-IDF is a numerical statistic that reflects the importance of a word in a document relative to a collection of documents (corpus)

```python
# Function to create TF-IDF features
def create_tfidf_features(X_train, X_test):
    tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=100000)
    xtrain_tfidf = tfidf_vect.fit_transform(X_train)
    xvalid_tfidf = tfidf_vect.transform(X_test)
    return xtrain_tfidf, xvalid_tfidf
```

```python
# Create TF-IDF features
xtrain_tfidf, xtest_tfidf = create_tfidf_features(X_train, X_test)
```

### Implementation and Evaluation of Machine Learning Model

```python
# Function for model training and evaluation
def train_and_evaluate_model(model, xtrain, ytrain, xtest, ytest):
    model.fit(xtrain, ytrain)
    train_predictions = model.predict(xtrain)
    test_predictions = model.predict(xtest)
    print("Confusion Matrix (Training Set):")
    plot_confusion_matrix(ytrain, train_predictions)
    print("\nClassification Report (Training Set):")
    print(classification_report(ytrain, train_predictions))
    print("\nAccuracy (Training Set):", accuracy_score(ytrain, train_predictions))
    print("\nConfusion Matrix (Testing Set):")
    plot_confusion_matrix(ytest, test_predictions)
    print("\nClassification Report (Testing Set):")
    print(classification_report(ytest, test_predictions))
    print("\nAccuracy (Testing Set):", accuracy_score(ytest, test_predictions))
```

```python
# Function to plot confusion matrix
def plot_confusion_matrix(true_labels, predicted_labels):
    cm = confusion_matrix(true_labels, predicted_labels)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()
```

### Initialize and fine-tune Random Forest classifier

```python
rf_classifier = RandomForestClassifier()
rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
```

```python
rf_grid_search = GridSearchCV(estimator=rf_classifier, param_grid=rf_param_grid, cv=3, n_jobs=-1)
# Fit the GridSearchCV object to the data
rf_grid_search.fit(xtrain_tfidf, y_train)

# Access the best estimator
best_rf_model = rf_grid_search.best_estimator_
```

```python
# Train and evaluate Random Forest model
print("Random Forest Model:")
train_and_evaluate_model(best_rf_model, xtrain_tfidf, y_train, xtest_tfidf, y_test)
```

### Initialize and fine-tune Support Vector Machine classifier

```python
svm_classifier = SVC()
svm_param_grid = {
    'C': [0.1, 1],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}
# Perform grid search
svm_grid_search = GridSearchCV(estimator=svm_classifier, param_grid=svm_param_grid, cv=3, n_jobs=-1)
svm_grid_search.fit(xtrain_tfidf, y_train)  # Fit grid search to training data

# Access the best estimator
best_svm_model = svm_grid_search.best_estimator_
```

```python
# Train and evaluate SVM model
print("\nSupport Vector Machine Model:")
train_and_evaluate_model(best_svm_model, xtrain_tfidf, y_train, xtest_tfidf, y_test)
```

# Deep Learning Modelling of the sentiment outcome
### Objective: To utilize deep learning algorithms for modeling the sentiment outcomes related to airline experiences.
### Approach: Implement various deep learning algorithms to model and predict sentiment outcomes in the context of airline customer experiences.
### Methodology: Employ algorithms such as Long Short-Term Memory (LSTM), and Bidirectional LSTM (BiLSTM) to train models on labeled data for sentiment prediction.

#### Tokenization function for better text preprocessing

```python

def tokenize_text(text):
    """
    Tokenizes the input text, converts tokens to lowercase alphanumeric characters,
    and joins them into a single string.
    
    Args:
        text (str): Input text to be tokenized.
        
    Returns:
        str: Processed text with tokens joined into a single string.
    """
    tokens = word_tokenize(text)  # Tokenize the text
    tokens = [token.lower() for token in tokens if token.isalnum()]  # Filter out non-alphanumeric tokens
    return ' '.join(tokens)  # Join tokens into a single string
```

```python
# Preprocess text data using the custom tokenizer function
X_train = X_train.apply(tokenize_text)
X_test = X_test.apply(tokenize_text)
```

```python
# Tokenize text data using Keras Tokenizer
tokenizer = Tokenizer(num_words=10000, oov_token='<OOV>')
tokenizer.fit_on_texts(X_train)
```

#### Encode labels using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

#### Convert text data to sequences and pad them

```python
maxlen = 100  # Maximum length of sequences
X_train_sequences = tokenizer.texts_to_sequences(X_train)
X_test_sequences = tokenizer.texts_to_sequences(X_test)

X_train_padded = pad_sequences(X_train_sequences, maxlen=maxlen, padding='post')
X_test_padded = pad_sequences(X_test_sequences, maxlen=maxlen, padding='post')
```

#### Define hyperparameters to tune and implementing the LSTM Model

```python
# Define hyperparameters to tune
embedding_dims = [64, 128]
lstm_units_list = [64, 128]
dropout_rates = [0.2, 0.3]

best_accuracy = 0.0
best_model = None

# Iterate over hyperparameter combinations
for embedding_dim in embedding_dims:
    for lstm_units in lstm_units_list:
        for dropout_rate in dropout_rates:
            # Build LSTM model with current hyperparameters
            model = Sequential()
            model.add(Embedding(input_dim=10000, output_dim=embedding_dim))
            model.add(SpatialDropout1D(dropout_rate))
            model.add(LSTM(units=lstm_units, dropout=dropout_rate, recurrent_dropout=dropout_rate))
            model.add(Dense(units=3, activation='softmax'))

            # Compile the model
            model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

            # Train the model
            history = model.fit(X_train_padded, y_train_encoded, epochs=10, batch_size=32, validation_split=0.2, verbose=0)

            # Evaluate the model on validation data
            _, val_accuracy = model.evaluate(X_test_padded, y_test_encoded, verbose=0)

            # Print current hyperparameters and validation accuracy
            print(f"Embedding Dim: {embedding_dim}, LSTM Units: {lstm_units}, Dropout Rate: {dropout_rate}, Validation Accuracy: {val_accuracy}")

            # Track the best model based on validation accuracy
            if val_accuracy > best_accuracy:
                best_accuracy = val_accuracy
                best_model = model
```

#### Evaluate the best model on test data for the LSTM Model which is one with the highest accuracy value
#### Embedding Dim: 64, LSTM Units: 64, Dropout Rate: 0.2 (Validity Accuracy: 0.3928)

```python
# Evaluate the best model on test data
y_pred_prob = best_model.predict(X_test_padded)
y_pred = np.argmax(y_pred_prob, axis=1)
```

```python
# Generate and plot confusion matrix
conf_matrix = confusion_matrix(y_test_encoded, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=label_encoder.classes_, 
            yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Convert label_encoder.classes_ to a list of strings
class_names = [str(cls) for cls in label_encoder.classes_]

# Print classification report with zero_division parameter
print(classification_report(y_test_encoded, y_pred, target_names=class_names, zero_division=0))
```

# Building and Modeling using BiLSTM Model

```python
# Build BiLSTM model
from keras.layers import Bidirectional
embedding_dim = 128
lstm_units = 128

model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=embedding_dim))
model.add(SpatialDropout1D(0.2))
model.add(Bidirectional(LSTM(units=lstm_units, dropout=0.2, recurrent_dropout=0.2)))
model.add(Dense(units=3, activation='softmax'))
```

#### Compile the BiLSTM model

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# Display model summary
model.summary()
```

### Define callbacks for learning rate scheduling and early stopping

```python
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.0001)
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

### Train the BiLSTM model with callbacks

```python
epochs = 20
batch_size = 32

history = model.fit(X_train_padded, y_train_encoded, epochs=epochs, batch_size=batch_size, 
                    validation_split=0.2, callbacks=[reduce_lr, early_stop])
```

### Evaluate the BiLSTM model on test data

```python
y_pred_prob = model.predict(X_test_padded)
y_pred = np.argmax(y_pred_prob, axis=1)
```

#### Generate and plot confusion matrix

```python
conf_matrix = confusion_matrix(y_test_encoded, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=label_encoder.classes_, 
            yticklabels=label_encoder.classes_)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

```

```python
# Convert label_encoder.classes_ to a list of strings
class_names = [str(cls) for cls in label_encoder.classes_]

# Print classification report with zero_division parameter
print(classification_report(y_test_encoded, y_pred, target_names=class_names, zero_division=0))
```
